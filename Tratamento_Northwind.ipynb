```python
order_details = pd.read_csv('/content/order_details.csv', delimiter=';')
order_details.head()
```

```python
# Creating collumn for total
order_details['total_amount'] = order_details['unit_price'] * order_details['quantity'] * (1 - order_details['discount'])
```

```python
order_details.head()
```

```python
orders = pd.read_csv('/content/orders.csv', delimiter=';')
orders.head()
```

```python
# Rename the column
orders.rename(columns={'ship_via': 'shipper_id'}, inplace=True)

orders.head()
```

```python
# Group order_details by 'order_id' and sum the 'total_amount'
order_total = order_details.groupby('order_id')['total_amount'].sum()

# Merge the aggregated totals back into orders
orders = orders.merge(order_total, on='order_id', how='left')

orders.head()
```

```python
products = pd.read_csv('/content/products.csv', delimiter=';')
products.head()
```

```python
df11 = pd.read_csv('/content/shippers.csv', delimiter=';')
df11.head()
```

```python
# Merge the 'orders' dataframe with 'df11' based on 'shipper_id'.
# The 'company_name' from 'df11' will be added to 'orders' as a new column named 'shipper_name'.
orders = pd.merge(orders, df11[['shipper_id', 'company_name']], on='shipper_id', how='left')

# Rename the 'company_name' column in the merged dataframe to 'shipper_name'.
orders.rename(columns={'company_name': 'shipper_name'}, inplace=True)
```
